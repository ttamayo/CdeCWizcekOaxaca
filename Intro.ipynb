A for loop goes over each element of a list, **note the spacing/indentation** after starting the for loop! For example, let's print all the elements in our a_list:

## 5.2 Conditional statements: If, else and else if

A conditional statement helps us to take a desition of whether the program is going to take or not certain actions.



```python
### For example, if we want to count how many times the word 'Chicken' appears in a_list:
counter = 0                ## Setting counter to 0
for i in a_list:
    if i == 'Chicken':
        counter += 1
print ('Chicken appeared: ',counter, 'times')

### Or if we want to check how many pixels has a 0.5 value in red:
reds = img[:,:,0]                            ## As before we are grabing just one chanel
reds_array = np.reshape(reds, (730*730))     ## As before, reshaping the channer red
counter = 0
for i in reds_array:
    if i > 0.5:
        counter += 1
print ('This number of pixels has a value bigger than 0.5 in red: ', counter)
```

# <i class="fa fa-check-circle"></i> Exercise: Find all pixels that have r > 0.8 and make them appear white

Imagine we were unable to see the color red, and so we want to encode "strong red signal" as "flickering white pixels"

```python
img_test_white = copy(img)  ## First we make a copy of our image
print (img_test_white.shape)
```

```python
## Later, we need to take a look into Loop through every pixel.
# If the red channel is above 0.8, make it white by turning all three channels to 1. 
length = 730
width = 730
for i in range(length):
    for j in range(width):
        pix_r = img_test_white[i,j,0]
        if pix_r > 0.8:
            img_test_white[i,j,0] = 1
            img_test_white[i,j,1] = 1
            img_test_white[i,j,2] = 1
            
#Show the resulting image
plt.imshow(img_test_white)
```

# <i class="fa fa-check-circle"></i> Exercise: Now make them black

```python
img_test_black = copy(img)
length = 730
width = 730
for i in range(length):
    for j in range(width):
        pix_r = img_test_black[i,j,0]
        if pix_r > 0.8:
            img_test_black[i,j,0] = 0
            img_test_black[i,j,1] = 0
            img_test_black[i,j,2] = 0

plt.imshow(img_test_black)
```

# 6. First temporal encoding exercise: make them oscillating white/black pixels: 

```python
### This function assembles frames into a gif.
def makemovie(n,moviename):
    images = []
    filenames = []
    for i in range(n):
        filenames.append("frame"+str(i)+".png")
    for filename in filenames:
        images.append(imageio.imread(filename))
    imageio.mimsave(moviename, images)
```

```python
num_frames = 40
for n in range(num_frames):
    if n%2 == 0:
        plt.imshow(img_test_black)
        pylab.savefig("frame"+str(n)+".png")
        
    else:
        plt.imshow(img_test_white)
        pylab.savefig("frame"+str(n)+".png")

moviename = 'movie1.gif'
makemovie(n,moviename)

def video(fname, mimetype):
    from IPython.display import HTML
    video_encoded = open(fname, "rb").read().encode("base64")
    video_tag = '<video controls alt="test" src="data:video/{0};base64,{1}">'.format(mimetype, video_encoded)
    return HTML(data=video_tag)

video("movie2.gif", "gif")
```
